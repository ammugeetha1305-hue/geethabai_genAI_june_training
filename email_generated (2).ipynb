```python
!pip install -q google-generativeai
import google.generativeai as genai
from google.colab import userdata
genai.configure(api_key=userdata.get('generate_ai'))

def generate_email(recipient, purpose, tone):
  user_prompt = f"""
  Write a professional email.

  Recipient: {recipient}
  Purpose: {purpose}
  Tone: {tone}
  IMPORTANT:
  - At the end of the email, sign off using:
    Name: Geetha Bai.S
    Designation: Team Lead

  Include:
  - Subject line
  - Greeting
  - Body
  - Closing
  """

  model = genai.GenerativeModel(
      model_name='gemini-2.5-flash-lite',
      system_instruction="You are a professional email writer who writes clear and structured emails."
  )

  response = model.generate_content(
      user_prompt,
      generation_config={
          "temperature": 0.3,
          "max_output_tokens": 300,
      }
  )

  print(response.text)

recipient = input("Enter recipient (e.g., HR Manager, TCS): ")
purpose = input("Enter purpose (follow-up / thank-you / leave request): ")
tone = input("Enter tone (formal / professional / friendly): ")

generate_email(recipient, purpose, tone)
```
