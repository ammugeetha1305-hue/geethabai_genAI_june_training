```python
!pip install -q google-generativeai

import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('gemini_api'))

knowledge_base = """
Store Name: Smart Home Appliances

=== Products ===

{
"question":"What products do you sell?",
"answer":"We sell TVs, Refrigerators, Washing Machines, Air Conditioners, Air Coolers, Water Heaters, Microwave Ovens, Mixers, Grinders, Fans, Chimneys, Water Purifiers and Kitchen Appliances."
},

{
"question":"Which TV brands are available?",
"answer":"We offer Samsung, LG, Sony, Xiaomi, TCL and Panasonic TVs."
},

{
"question":"Which refrigerator brands are available?",
"answer":"Samsung, LG, Whirlpool, Haier and Godrej refrigerators are available."
},

{
"question":"Do you sell inverter ACs?",
"answer":"Yes, we have inverter and non-inverter air conditioners from leading brands."
},

{
"question":"Do you sell fully automatic washing machines?",
"answer":"Yes, both top-load and front-load fully automatic washing machines are available."
},

=== Prices ===

{
"question":"What is the price of a Smart TV?",
"answer":"Smart TVs start from Rs.18,000 depending on the brand and size."
},

{
"question":"What is the price of a refrigerator?",
"answer":"Refrigerators start from Rs.15,000."
},

{
"question":"What is the price of an air conditioner?",
"answer":"Air conditioners start from Rs.30,000."
},

{
"question":"Do you have any discounts?",
"answer":"Yes. Seasonal offers and festival discounts are available on selected products."
},

=== EMI & Payment ===

{
"question":"Do you provide EMI?",
"answer":"Yes. EMI options are available through major banks and finance partners."
},

{
"question":"Can I pay using UPI?",
"answer":"Yes. UPI, Debit Card, Credit Card, Cash and Net Banking are accepted."
},

{
"question":"Is zero-cost EMI available?",
"answer":"Yes. Zero-cost EMI is available on selected products and bank cards."
},

=== Delivery ===

{
"question":"Do you provide home delivery?",
"answer":"Yes. Free home delivery is available within the city for eligible products."
},

{
"question":"How many days does delivery take?",
"answer":"Delivery usually takes 1 to 3 working days depending on stock availability."
},

{
"question":"Do you install the products?",
"answer":"Yes. Installation is available for eligible products through authorized service partners."
},

=== Warranty ===

{
"question":"What warranty do products have?",
"answer":"Warranty depends on the manufacturer. Most products include a 1-year manufacturer warranty."
},

{
"question":"Can I extend my warranty?",
"answer":"Yes. Extended warranty plans are available for selected products."
},

=== Exchange ===

{
"question":"Do you accept old appliances?",
"answer":"Yes. Exchange offers are available for selected products."
},

{
"question":"Can I exchange my old TV?",
"answer":"Yes. Old TVs can be exchanged based on condition and eligibility."
},

=== Service ===

{
"question":"How do I book a service?",
"answer":"Contact our customer support or visit the nearest showroom to register a service request."
},

{
"question":"How do I track my order?",
"answer":"Provide your Order ID to customer support for order status."
},

=== Contact ===

Customer Care:
Phone: 1800-111-222
Email: support@smarthome.com

Working Hours:
Monday to Saturday
9:00 AM to 8:00 PM
"""

system_instruction = f"""
You are a friendly customer support chatbot for Smart Home Appliances.

Rules:

1. Answer ONLY from the knowledge base.
2. If the answer is unavailable, reply exactly:
"I'm sorry, I don't have that information. Please contact Customer Care at 1800-111-222 or support@smarthome.com."
3. Be polite.
4. Keep answers short.
5. Never make up information.

Knowledge Base:

{knowledge_base}
"""

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash-lite",
    system_instruction=system_instruction
)

chat = model.start_chat(history=[])

print("🏠 Smart Home Appliances Support Bot is Ready!")
print("Type 'quit' to exit.\n")

while True:

    user_input = input("You: ").strip()

    if not user_input:
        continue

    if user_input.lower() in ["quit","exit","bye"]:
        print("Bot: Thank you for visiting Smart Home Appliances. Have a great day! 😊")
        break

    response = chat.send_message(user_input)
    print("Bot:", response.text)
    print()
```
